```python
from argparse import Namespace
from pathlib import Path
import pandas as pd

args = Namespace(
    results_dir=Path("/home/jupyter/linnea_results/"),
    ablation_base_name="ioncastgnn-fulldataset-dilation128-residual-oct22-technical-showcase-ablation-",
    ablations=["celestrak", "f107", "noaux", "not-residual-target", "omni", "qd-sunmoon", "set"],# "solarmax", "solarmin"],
    epochs=8,
    long_horizon="long-horizon"
    )
```

```python
values_long_horizon = {}
values_lead = {}
values_180_lead_time = {}
for ablation in args.ablations:
    ablation_dir = Path(args.ablation_base_name + ablation)
    filename_long = Path(f"epoch-{args.epochs:02}-valid-long-horizon-metrics.csv")
    filename_lead = Path(f"epoch-{args.epochs:02}-valid-fixed-lead-time-metrics.csv")
    file_path_long = args.results_dir / ablation_dir / filename_long  
    file_path_lead = args.results_dir / ablation_dir / filename_lead  
    # print(file_path_long)
    # print("/home/jupyter/linnea_results/ioncastgnn-fulldataset-dilation128-residual-oct22-technical-showcase-ablation-celestrak/epoch-08-valid-long-horizon-metrics.csv")
    # print(file_path_lead)

    df_long = pd.read_csv(file_path_long)
    df_lead = pd.read_csv(file_path_lead)
    print(df_lead[df_lead['event_category'] == 'all'])
    values_long_horizon[ablation] = [df_long[df_long['event_id'] == 'all-mean']["jpld_unnormalized_rmse"].iloc[0].item(), df_long[df_long['event_id'] == 'all-std']["jpld_unnormalized_rmse"].iloc[0].item()]
    values_lead[ablation] = [df_lead[df_lead['event_category'] == 'all']["mean_rmse_tecu"].iloc[0].item(),              df_lead[df_lead['event_category'] == 'all']["std_rmse_tecu"].iloc[0].item()]
    # if df_long[df_long['event_id'] == 'all-mean']:
    #     print(ablation, df_long['jpld_rmse'].iloc[0])
    # print(df_long['event_id'].iloc[0] == 'all-mean', ablation)
```

```python
from pprint import pprint
print("Long horizon RMSE (mean, std):")
pprint(values_long_horizon)
print()
print("180 min lead time RMSE (mean, std):")
pprint(values_lead)
```
