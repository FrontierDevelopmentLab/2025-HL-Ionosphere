```python
import pandas as pd
```

```python
file_name_original = '/disk2-ssd-8tb/data/2025-hl-ionosphere//omniweb_karman_2025/merged_omni_solar_wind.csv'

date_start = '2010-05-13T00:00:00'
date_end = '2024-08-01T00:00:00'

# Convert the date strings to pandas datetime objects
date_start = pd.to_datetime(date_start)
date_end = pd.to_datetime(date_end)

# Process data to be within a starting and ending data and also only keep data with minutes :00, :15, :30, :45
def process_data(data, date_start, date_end):
    data['all__dates_datetime__'] = pd.to_datetime(data['all__dates_datetime__'], format='%Y-%m-%d %H:%M:%S')
    data = data[(data['all__dates_datetime__'] >= date_start) & (data['all__dates_datetime__'] <= date_end)]
    data = data[data['all__dates_datetime__'].dt.minute.isin([0, 15, 30, 45])]
    return data

data = pd.read_csv(file_name_original)
# print data summary, like number of rows and columns
print(f"Data shape: {data.shape}")
print(f"Columns: {data.columns.tolist()}")

data2 = process_data(data, date_start, date_end)
print(f"Processed data shape: {data2.shape}")
print(f"Processed data columns: {data2.columns.tolist()}")

# how many rows the data should be between date_start and date_end, every 15 minutes
expected_rows = (date_end - date_start).total_seconds() / 900  # 15 minutes in seconds
print(f"Expected rows: {expected_rows}")
print(f"Actual rows: {data2.shape[0]}")
```

```python
file_name_new = file_name_original.replace('.csv', '_processed.csv')
data2.to_csv(file_name_new, index=False)
print(f"Processed data saved to: {file_name_new}")
```
