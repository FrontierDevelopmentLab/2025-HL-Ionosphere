```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm

# Example data
T, H, W = 4, 64, 128
arr = qd_sample

fig, ax = plt.subplots()
offset = 20  # shift between frames

for i in range(T):
    # shift each frame diagonally by i*offset
    extent = [i*offset, W + i*offset, i*offset, H + i*offset]
    ax.imshow(arr[i], cmap=cm.viridis, extent=extent)

ax.set_xlim(0, W + (T-1)*offset + 20)
ax.set_ylim(0, H + (T-1)*offset + 20)
plt.show()
```
